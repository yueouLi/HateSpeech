| **id** | **example** | **term** | **pos** | **sense_id** | **definition** | **ANNOTATION** | **DEFICIENT SENTENCE** |

As the default setting, the color of the definition text is white to make it invisible. 

**TASK**

| Label |                  |
|-------|------------------|
| 0     | Not hateful      |
| 1     | Weakly hateful   |
| 2     | Strongly hateful |
| -     | Cannot decide    |
