```python
from pigeon import annotate
import pandas as pd

# 读取Excel文件
orig_sentence_file = "gaze4hate_sentences_to_annotate.xlsx"
orig_sent_df = pd.read_excel(orig_sentence_file, sheet_name='stimuli')
orig_sent_df = orig_sent_df.loc[orig_sent_df['Cond_remark'] != 'fam']

# 从DataFrame中提取句子列表
sentences = orig_sent_df["processed_assertion"].tolist()

# 使用PIGEON进行注释
annotations = annotate(
    sentences, 
    options=["implicitly", "explicitly", 'undefined']
)

df = pd.DataFrame(annotations, columns=['Annotation'])
df.to_csv('annotations.csv',index=False )

# 创建包含句子和标注内容的DataFrame
```

```python
annotations
```

```python
with open('annotations.txt', 'w') as f:
    for annotation in annotations:
        # 将每个注释转换为字符串并写入文件
        f.write(str(annotation) + '\n')
```

```python
import pandas as pd

# 读取CSV文件
orig_sentence_file = "gaze4hate_sentences_to_annotate.xlsx"
orig_sent_df = pd.read_excel(open(orig_sentence_file, 'rb'), sheet_name='stimuli') 
# 初始化token和types的集合
tokens = set()
types = set()

# 遍历每个句子
for sentence in orig_sent_df["processed_assertion"]:
    # 分割句子成token
    sentence_tokens = sentence.split()
    
    # 将token添加到tokens集合中
    tokens.update(sentence_tokens)
    
    # 将token添加到types集合中（去重）
    types.update(set(sentence_tokens))

# 计算总的token数和types数
total_tokens = sum(len(sentence.split()) for sentence in orig_sent_df["processed_assertion"])
total_types = len(types)

print(f"Total tokens: {total_tokens}")
print(f"Total types: {total_types}")
```

```python

orig_sentence_file = "gaze4hate_sentences_to_annotate.xlsx"
orig_sent_df = pd.read_excel(open(orig_sentence_file, 'rb'), sheet_name='stimuli') 

# 统计每个句子中的单词数量
word_counts = orig_sent_df["processed_assertion"].apply(lambda x: len(x.split()))

# 打印每个句子的单词数量
for index, word_count in enumerate(word_counts, start=1):
    print(f"Sentence {index}: {word_count} words")

# 统计总单词数量
total_words = word_counts.sum()
print(f"Total words: {total_words}")



# 计算最长句子、最短句子和平均每个句子的单词数量
max_words = word_counts.max()
min_words = word_counts.min()
average_words = word_counts.mean()

# 打印结果
print(f"Longest sentence: {max_words} words")
print(f"Shortest sentence: {min_words} words")
print(f"Average words per sentence: {average_words:.2f} words")
```
